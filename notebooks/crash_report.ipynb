```python
weather=df['weather'].unique()
```

```python
count=list()
for i in weather:
    count.append(df[df['weather']==i]['weather'].count())
wt_count=pd.DataFrame({'weather':weather,'count':count})
wt_count.dropna(inplace=True)
```

```python
wt_count
```

```python
wt_count['perc']=wt_count['count']/wt_count['count'].sum() *100
```

```python
sns.barplot(data=wt_count,y='weather',x='count',width=0.5)
plt.title('Count of crashes in weather conditions')
```

```python
sns.barplot(data=wt_count,y='weather',x='perc',width=0.5)
plt.title('Percentage of crashes in weather conditions')
```
