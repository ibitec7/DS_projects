```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import datatools as dt
```

```python
x = np.array([2,9,3,4,7])
```

```python
s = np.std(x)
x_hat = np.mean(x)
```

```python
print(f"Standard Deviation: {s: .10f} \nMean: {x_hat: .10f}")
```

```python
sns.histplot(x)
plt.title("Histogram of variable X")
plt.xlabel("x")
plt.ylabel("Frequency f")
plt.axvline(x_hat,linestyle='--',label="Mean",color="red")
sns.ecdfplot(x, color='red',stat="proportion",label="CDF plot")
plt.grid(linestyle="--")
plt.legend()
plt.show()
```
